```python
import numpy as np
import math
import matplotlib.pyplot as plt

def GAUSS(x, y, xSample):
  F = 0
  for i in range(len(y)):
    p = 1
    for j in range(len(x)):
      if i != j:
        p = p * (xSample - x[j]) / (x[i] - x[j]) #    t
    F = F + p * y[i]
  return F

def F(x):
  return math.cos(math.cos(5 * x))
  # return x * math.exp(x) + math.sin(x)

def R(y, yInters):
  arr = []
  for i in range(len(y)):
    arr.append(abs(y[i] - yInters[i]))
  arr = np.array(arr)
  return arr

X = np.array(np.arange(-2, 0.1, 0.1), dtype=float)
Y = np.array([F(x) for x in X])

x = np.array(np.arange(-2, 0.1, 1), dtype=float)
G2 = np.array([F(x) for x in x])

fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (20, 10))
ax1, ax2 = axes

nodes = 21

rY_G2 = []
yInters = []

for x, y in [[x, G2]]:
  xSamples = np.linspace(np.min(x), np.max(x), nodes)
  yInters = [GAUSS(x, y, i) for i in xSamples]
  ax1.plot(xSamples, yInters, color='r')

rY_G2 = R(Y, yInters)

for x, y in [[X, Y]]:
  ax1.plot(x, y, color='b')
  ax1.plot(x, y, marker='o', color='b', ls='')

for x, y in [[X, rY_G2]]:
  ax2.plot(x, y, color='b')

plt.grid(True)
plt.show()
```
