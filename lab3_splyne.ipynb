```python
import numpy as np
import math
import matplotlib.pyplot as plt

def SPL(x, y):
  arr = []
  for i in range(10,81,10):
    for j in range(10):
      arr.append(y[i] + (x[i] - x[j + (i - 10)]) * (y[i - 10] - y[i]) / (x[10] - x[0]))
  arr.append(y[80] + (x[80] - x[80]) * (y[70] - y[80]) / (x[10] - x[0]))
  return arr

def F(x):
  return math.cos(math.cos(5 * x))
  # return x * math.exp(x) + math.sin(x)

def R(y, yInters):
  arr = []
  for i in range(len(y)):
    arr.append(abs(y[i] - yInters[i]))
  arr = np.array(arr)
  return arr

X = np.array(np.arange(-2, 0.001, 0.025), dtype=float)
Y = np.array([F(x) for x in X])


fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (20, 10))
ax1, ax2 = axes

nodes = 81

rY_G2 = []
yInters = []

for x, y in [[X, Y]]:
  xSamples = np.linspace(np.min(x), np.max(x), nodes)
  yInters = SPL(x, y)
  ax1.plot(xSamples, yInters, color='r')
  ax1.plot(xSamples, yInters, marker='.', color='r')

rY_G2 = R(Y, yInters)

for x, y in [[X, Y]]:
  ax1.plot(x, y, color='b')

for x, y in [[X, rY_G2]]:
  ax2.plot(x, y, color='b')

plt.grid(True)
plt.show()

# 
```
